## Series

Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index. The basic method to create a Series is to call: \\
```python
import pandas as pd
s = pd.Series(data, index=index) 
```

data can be many different things: 
- Python dict
- ndarray
- simple scalar value 

The passed index is a list of axis labels.

## Vectorized operations and label alignment with Series
- When working with raw NumPy arrays, looping through value-by-value is usually not necessary. 
- The same is true when working with Series in pandas. Series can also be passed into most NumPy methods expecting an ndarray.

````python
s
````

````python
s+s
````

````python
np.exp(s)
````

A key difference between Series and ndarray is that operations between **Series** automatically align the data based on label. Thus, you can write computations without giving consideration to whether the Series involved have the same labels.

````python
s[1:]
````

````python
s[:-1]
````

````python
s[1:] + s[:-1]
````

````python
s[1:].to_numpy() + s[:-1].to_numpy()
````

If unaligned, the result will be marked as missing NaN (or missing value will be considered as NaN value)

## DataFrame

DataFrame is a 2D labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object. Like Series, DataFrame accepts many different kinds of input:

- Dict of 1D ndarrays, lists, dicts, or Series
- 2-D numpy.ndarray
- Structured or record ndarray
- A Series
- Another DataFrame

### From dict of Series or dicts

````python
d = {
    "one": pd.Series([1.0, 2.0, 3.0], index=["a", "b", "c"]),
    "two": pd.Series([1.0, 2.0, 3.0, 4.0], index=["a", "b", "c", "d"]),
}
df = pd.DataFrame(d)
df
````

You can specify the index (row label)

````python
pd.DataFrame(d, index=["d", "b", "a"])
````

You can specify the column (column label)\
since defined from dict (key value: one,two), will only call "two", leaving "three" column as all NaN values

````python
pd.DataFrame(d, index=["d", "b", "a"], columns=["two", "three"])
````

````python
df.index
````

````python
df.columns
````

### Indexing / selection

df[col] : returns Series

df.loc[label]  returns Series (select row by label)

df.iloc[loc] returns Series (select row by integer location)

df[5:10] returns empty dataframe

````python
df = pd.DataFrame(data=[[x,x] for x in range(1,5)],columns=['bar','one'],\
                  index=list('abcd'))
df
````
