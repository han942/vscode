## squeeze

remove axes of length one from input

```python
t = np.arange(6)
t.shape
```

```python
t = np.reshape(t,(6,-1))
print(t)
t.shape #2D array
```

```python
display(np.squeeze(t.reshape(-1,6)))
np.squeeze(t.reshape(3,2))
```

```python
print(np.squeeze(t))
np.squeeze(t).shape
```

## stack

Join a sequence of arrays along a new axis.

```python
array = np.random.randn(3, 4) 
print(array.shape)
array
```

```python
arrays = [np.random.randn(3, 4) for _ in range(10)]
arrays
```

```python
np.stack(arrays, axis=0).shape
```

```python
np.stack(arrays, axis=1).shape
```

```python
np.stack(arrays, axis=2).shape
```

```python
x = np.array([1,2,3,4]).reshape(2,2)
y = np.arange(5,9).reshape(2,2)

np.stack((x,y),axis=-1)
np.vstack((x,y))
```

## vstack

stack arrays in sequence vertically (row wise)

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
np.stack((a,b),axis=1)
```

```python
np.vstack((a,b))
```

```python
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])
```

```python
np.vstack((a,b))
```

## hstack

Stack arrays in sequence horizontally (column wise)

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
np.stack((a,b),axis=0)
```

```python
np.hstack((a,b))
```

```python
a = np.array([[1],[2],[3]])
b = np.array([[4],[5],[6]])
```

```python
np.hstack((a,b))
```

## Split

Split an array into multiple sub-arrays as views into array.

```python
x = np.arange(9.0)
```

```python
np.split(x, 3)
```

```python
np.split(x, 4) # Not evenly distributed 
```

```python
x = np.arange(8.0)
np.split(x, [3, 5, 6, 10])
```

## flip

Reverse the order of elements in an array along the given axis.

The shape of the array is preserved, but the elements are reordered.

```python
A = np.arange(8).reshape((2,2,2))
A
```

```python
np.flip(A,axis=0)
```

```python
np.flip(A,1)
```

```python
np.flip(A,axis=2)
```

## trim_zeros

Trim the leading and/or trailing zeros from a 1-D array or sequence.

```python
a = np.array((0, 0, 0, 1, 2, 3, 0, 2, 1, 0))
```

```python
np.trim_zeros(a) # default: trimming both sides
```

```python
np.trim_zeros(a,'f') # fowrard trimming
```

```python
np.trim_zeros(a,'b') # backward trimming
```

```python
np.ma.masked_equal(a,0) # masking the matching values
```
