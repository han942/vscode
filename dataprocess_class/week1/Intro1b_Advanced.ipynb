# Class

Classes provide a means of bundling data and functionality together. Creating a new class creates a new type of object, allowing new instances of that type to be made. Each class instance can have attributes attached to it for maintaining its state. Class instances can also have methods (defined by its class) for modifying its state.

- Class - Creates new type of object (making my own type) / Can bundle data (also function)
- Class Instance
- **Class constructor** : __init__ called when initializing the given func(class)

- self.variable (인스턴스 변수): allocates each data storage for the instance. 

```python
class Calculate:
    def __init__(self, first, second):  # constructor 
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return 'summation : %d' % result  
    
    def multiply(self):
        result = self.first * self.second
        return 'multiplication : %d' % result  
    
myinstance = Calculate(5,6) # creating instance -> assigning (first,second) to the given variables
print(myinstance)

print(myinstance.add())
```

## practice for multiplication

```python
print(myinstance.multiply())
```

## class constructor

- Can initialize number within the __init__ method

```python
class DemoClass:
    # constructor
    def __init__(self):
        # initializing instance variable
        self.num=100

    # a method
    def read_number(self):
        print(self.num)


# creating object of the class. This invokes constructor
obj = DemoClass()

# calling the instance method using the object obj
obj.read_number()
```

Default constructor 

```python
class DemoClass:
    num = 100

    # a method
    def read_number(self):
        print(self.num)


# creating object of the class
obj = DemoClass()

# calling the instance method using the object obj
obj.read_number()
```

Constructor Declaration

```python
class DemoClass:
    num = 101

    # non-parameterized constructor
    def __init__(self):
        self.num = 999

    # a method
    def read_number(self):
        print(self.num)


# creating object of the class
obj = DemoClass()

# calling the instance method using the object obj
obj.read_number()
```

```python
class DemoClass:
    num = 101

    # parameterized constructor
    def __init__(self, data):
        self.num = data

    #  method
    def read_number(self):
        print(self.num)


# creating object of the class
# this will invoke parameterized constructor
obj = DemoClass(55)

# calling the instance method using the object obj
obj.read_number()

# creating another object of the class
obj2 = DemoClass(66)

# calling the instance method using the object obj
obj2.read_number()
```

## class inheritance
: Containing all of the parent attributes (methods)

```python
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print(self.firstname, self.lastname)

#Use the Person class to create an object, and then execute the printname method:

x = Person("John", "Doe")
x.printname()
```

inheriting parent class

```python
class Student(Person):
    pass

x = Student("John", "Cho")
x.printname()
```

```python
class Student(Person):
    def __init__(self, fname, lname, year):
        super().__init__(fname, lname) # inherent the initialization method of parent's class
        self.graduationyear = year
    def printall(self):
        print(self.firstname, self.lastname, self.graduationyear)
```

```python
x = Student("John", "Cho", 2027)
x.printname()
```

```python
x.printall()
```

```python
class Student(Person):
    def __init__(self, fname, lname, year=2027):
        super().__init__(fname, lname)
        self.graduationyear = year
    def printall(self):
        print(self.firstname, self.lastname, self.graduationyear)
        
x = Student("John", "Cho")
x.printname()
x.printall()
```

## method overriding 
**[def]**: refining the inherited method for different purpose compared with the parent's method in child's class
- Essential for programming
- The priority for method __init__ focuses on the child's method
    - redefinement is possible for inhereted methods given by the parent's class

```python
class Parent():
      
    # Constructor
    def __init__(self):
        self.value = "Inside Parent"
          
    # Parent's show method
    def show(self):
        print(self.value)

# Defining child class
class Child(Parent):
    pass
```

```python
# Driver's code
obj1 = Parent()
obj2 = Child() # the initializer correspondingly used in the child's class
obj1.show()
obj2.show()
```

```python
class Parent():
      
    # Constructor
    def __init__(self):
        self.value = "Inside Parent"
          
    # Parent's show method
    def show(self):
        print(self.value)
          
# Defining child class
class Child(Parent):
      
    # Constructor diff with the parent class
    def __init__(self):
        self.value = "Inside Child"
          
    # Child's show method
    def show(self):
        print("def",self.value)
```

```python
# Method overriding -> prioritizes the nearest class (or declaring the inhereted method inside the child class)
obj1 = Parent()
obj2 = Child()
  
obj1.show()
obj2.show()
```

# method overloading

Two or more methods have the same name but different numbers of parameters or different types of parameters, or both. These methods are called overloaded methods and this is called method overloading. 

```python
def product(a, b):
    p = a * b
    print(p)

def product(a,b,c):
    p = a*b*c
    print(p)

product(4,5)
```

```python
# Function to take multiple arguments
def add(datatype, *args):   # Considering *args like tuple(can put as much as you want) - able by putting &

    # if datatype is int
    # initialize answer as 0
    if datatype == 'int':
        answer = 0
 
    # if datatype is str
    # initialize answer as ''
    if datatype == 'str':
        answer = ''
 
    # Traverse through the arguments
    for x in args:
 
        # This will do addition if the args are int.
        # Or concatenation if the arguments are str
        answer = answer + x
 
    print(answer)
    
add('int', 5,6,7,8,9)
add('str', 'Hi ', 'Geeks',' This',' is') # 'Hi' +'Geeks' + 'This' + 'is'
add('str',0,'0',123)
```

### More Arguments

- 'a' doesn't have star marks, which considers as a fixed parameter
- declaring the *arg always should be the last part of the given input

- 'd' is always fixed as the value 4 
-------------------------------

- d=4 (default args) : args that are defined with the given value if not called (@ its position)
- 'a','c' (Positional args) : args that needs to be defined (may cause error if not)
- \* (variable length args, 가변 위치 인자): gathers remaining positional args with Tuple
- ** (keyword args, **kwargs) : given with multiple {key-values}, declared with dictionary

```python
def foo(a, *arg):
    return a, arg

print(foo(1,2,3))
print(foo(1,2,3,4))
```

```python
def foo(a, c, d=4, *arg, **karg):
    return a, c, d, arg, karg
foo(1,2)
```

```python
foo(1, 2, 3, 4, 5) # d가 비어있지 않아서 3으로 들어감
```

```python
foo(1, 2, 3, 5,6,7, h=2, f=3, j=3)
```

```python
foo(1,2,5,6,h=2,f=3,j=3)
```
