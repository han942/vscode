```python
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np

img = misc.face()
print('Maximum value of image: ',img.max())
img_array = img / 255 # Normalization

red_array = img_array[:, :, 0]
green_array = img_array[:, :, 1]
blue_array = img_array[:, :, 2]

img_gray = img_array @ [0.2126, 0.7152, 0.0722]
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
#Full-SVD
U,s,Vt = np.linalg.svd(img_gray)
U.shape,s.shape,Vt.shape
```

```python
Sigma = np.zeros(shape=(U.shape[1],Vt.shape[0]))
np.fill_diagonal(Sigma,s)
```

```python
np.linalg.norm(img_gray - U @ Sigma @ Vt)
```

```python
plt.plot(s) # at svd, values of s are sorted in descending order
```

```python
#Truncated-SVD
k = 25
approx = U @ Sigma[:,:k] @ Vt[:k,:]
```

```python
np.linalg.norm(img_gray - approx)
```
